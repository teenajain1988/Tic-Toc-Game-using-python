```python
def displayboard(arr):
    print(arr[1]+"|"+arr[2]+"|"+arr[3])
    print(arr[4]+"|"+arr[5]+"|"+arr[6])
    print(arr[7]+"|"+arr[8]+"|"+arr[9])
```

```python
def decide_marker():
    marker = ""
    while marker != "X" and marker != "O":
        marker = input("decide your mark : X/O  ").upper()
    player1 = marker
    if player1 == "X":
        player2 = "O"
    else:
        player2 = "X"
    
    return player1,player2
        
    
    
```

```python
def choosefirst():
    from random import randint
    turn = randint(0,1)
    print(turn)
    if turn == 1:
        return player1
    else:
        return player2
    
```

```python
def blank_space(arr):
    for i in range(1,10):
        if arr[i] == " ":
            return True
            break
    else:
        return False
```

```python
def place_marker(arr,marker):
    position = int(input("enter the position from 1 to 9:  "))
    arr[position] = marker
```

```python
def win_check(arr,mark):
    return (arr[1] == mark and arr[2] == mark and arr[3] == mark or
    arr[4] == mark and arr[5] == mark and arr[6] == mark or
    arr[7] == mark and arr[8] == mark and arr[9] == mark or
    arr[1] == mark and arr[4] == mark and arr[7] == mark or
    arr[2] == mark and arr[5] == mark and arr[8] == mark or
    arr[3] == mark and arr[6] == mark and arr[9] == mark or
    arr[1] == mark and arr[5] == mark and arr[9] == mark or
    arr[3] == mark and arr[5] == mark and arr[7] == mark)

     
    
    
```

```python
def fullboardcheck(arr):
    for i in range(1,10):
        if arr[i] == " ":
            return False
            break
    else:
        return True
        
    
```

```python
def replay():
    re = input("do u want to replay: Y/N    ").upper()
    return re == "Y"
    
```

```python
while True:
    game_on = True
    arr = [" "]*10
    print("welcome to tic toc toe")
    choice = input("are you ready to play: Y/N    ").upper()
    print(choice)
    if choice == "Y":
        displayboard(arr)
        player1,player2 = decide_marker()
        print(f" player1 is {player1}")
        print(f" player2 is {player2}")
        turn = choosefirst()
        print(f" turn no is {turn}")
        while game_on:
            if turn == player1:
                if blank_space(arr):
                    place_marker(arr,player1)
                    displayboard(arr)
                if win_check(arr,player1):
                    print("congratulation player is  the winner")
                    game_on = False
                else:
                    if fullboardcheck(arr):
                        print("Tie")
                        game_on = False
                    else:
                        turn = player2
            if turn == player2:
                if blank_space(arr):
                    place_marker(arr,player2)
                    displayboard(arr)
                if win_check(arr,player2):
                    print("congratulation u means player2 is winner")
                    game_on = False
                else:
                    if fullboardcheck(arr):
                        print("Tie")
                        game_on = False
                    else:
                        turn = player1
    if not replay():
        break
        
        

    

    
```
